```python
X=data1.drop(['patientid','target'],axis=1)
X.columns

y=data1['target']
patientid=data1['patientid']
```

```python
from sklearn.model_selection import train_test_split

# Divide os dados em treino (80%) e teste (20%)
X_train, X_test, y_train, y_test, id_train, id_test = train_test_split(
    X,          # Features
    y,          # Target (DoenÃ§a CardÃ­aca)
    patientid,  # IDs dos pacientes
    test_size=0.2,      # 20% para teste
    random_state=42     # Semente para reprodutibilidade
)

from sklearn.linear_model import LogisticRegression  # Modelo
from sklearn.metrics import accuracy_score, classification_report  # MÃ©tricas
```

```python
# Importar normalizador
from sklearn.preprocessing import StandardScaler

# Criar objeto de normalizaÃ§Ã£o
scaler = StandardScaler()

# Normalizar dados de treino (fit + transform)
X_train = scaler.fit_transform(X_train)

# Normalizar dados de teste (apenas transform)
X_test = scaler.transform(X_test)
```

```python
# hyperparameters = {
#     'penalty': 'l1',           # RegularizaÃ§Ã£o Lasso
#     'solver': 'liblinear',     # Algoritmo de otimizaÃ§Ã£o
#     'max_iter': 1000          # NÃºmero mÃ¡ximo de iteraÃ§Ãµes
# }

hyperparameters = {
    'penalty': 'elasticnet',   # RegularizaÃ§Ã£o combinada
    'solver': 'saga',          # Algoritmo de otimizaÃ§Ã£o
    'l1_ratio': 0.5,          # Mistura de L1 e L2
    'max_iter': 1000          # NÃºmero mÃ¡ximo de iteraÃ§Ãµes
}
```

```python
# Criar e treinar o modelo de regressÃ£o logÃ­stica
model_log = LogisticRegression(**hyperparameters)  # Instancia o modelo com hiperparÃ¢metros
result_log = model_log.fit(X_train, y_train)      # Treina o modelo

# Fazer prediÃ§Ãµes no conjunto de teste
y_log_pred = model_log.predict(X_test)            # Gera prediÃ§Ãµes

# Calcular e mostrar acurÃ¡cia
accuracy = accuracy_score(y_test, y_log_pred)      # Calcula precisÃ£o
print(f"PrecisÃ£o: {accuracy:.2f}")                 # Mostra resultado em formato decimal
```

```python
# Gerar e imprimir relatÃ³rio de classificaÃ§Ã£o
print("RelatÃ³rio de ClassificaÃ§Ã£o:")
print(classification_report(y_test, y_log_pred))
```

```python
# Importar mÃ©tricas do sklearn
from sklearn import metrics

# Criar matriz de confusÃ£o
confusion_matrix = metrics.confusion_matrix(y_test, y_log_pred)
```

```python
# Importar statsmodels
import statsmodels.api as sm

# Criar modelo de regressÃ£o logÃ­stica
logit_model = sm.Logit(y_train, X_train)  # Define modelo com dados de treino

# Treinar o modelo
result_smlog = logit_model.fit()  # Ajusta o modelo aos dados

# Imprimir sumÃ¡rio estatÃ­stico detalhado
print(result_smlog.summary())
```

```python
# Realizar prediÃ§Ãµes com o modelo treinado
smlog_pred = result_smlog.predict(X_test)  # Gera probabilidades

# Converter probabilidades em classes binÃ¡rias (0 ou 1)
smlog_pred_binary = (smlog_pred >= 0.5).astype(int)  
# Se probabilidade >= 0.5: classe 1 (doenÃ§a)
# Se probabilidade < 0.5: classe 0 (sem doenÃ§a)

# Calcular e mostrar acurÃ¡cia
accuracy_logit = accuracy_score(y_test, smlog_pred_binary)
print("AcurÃ¡cia:", accuracy_logit)
```

```python
# Importar mÃ©tricas para anÃ¡lise ROC
from sklearn.metrics import roc_curve, roc_auc_score

# Calcular score AUC
auc_score = roc_auc_score(y_test, smlog_pred_binary)
```

```python
# Calcular pontos da curva ROC
fpr, tpr, thresholds = roc_curve(y_test, smlog_pred_binary)

# Criar grÃ¡fico
plt.figure(figsize=(8, 6))

# Plotar curva ROC
plt.plot(fpr, tpr, label=f'Curva ROC (AUC = {auc_score:.2f})')

# Plotar linha diagonal de referÃªncia
plt.plot([0, 1], [0, 1], linestyle='--', color='r')

# Configurar rÃ³tulos e tÃ­tulo em portuguÃªs
plt.xlabel('Taxa de Falsos Positivos (1-Especificidade)')
plt.ylabel('Taxa de Verdadeiros Positivos (Sensibilidade)')
plt.title('Curva ROC para RegressÃ£o LogÃ­stica')

# Adicionar legenda e grade
plt.legend()
plt.grid(True)

# Mostrar grÃ¡fico
plt.show()
```

```python
# Importar bibliotecas necessÃ¡rias
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Calcular matriz de confusÃ£o
cmlg = confusion_matrix(y_test, smlog_pred_binary)

# Criar figura
plt.figure(figsize=(8, 6))

# Criar heatmap da matriz de confusÃ£o
sns.heatmap(cmlg, 
            annot=True,              # Mostrar valores
            cmap='Blues',            # Esquema de cores
            fmt='g',                 # Formato dos nÃºmeros
            cbar=False)             # Remover barra de cores

# Configurar rÃ³tulos em portuguÃªs
plt.xlabel('PrevisÃµes')
plt.ylabel('Valores Reais')
plt.title('Matriz de ConfusÃ£o - Dados de Teste')

# Adicionar rÃ³tulos das classes
plt.xticks([0.5, 1.5], ['Sem DoenÃ§a', 'Com DoenÃ§a'])
plt.yticks([0.5, 1.5], ['Sem DoenÃ§a', 'Com DoenÃ§a'])

# Mostrar grÃ¡fico
plt.show()
```

```python
# 1. ReduÃ§Ã£o de Dimensionalidade
from sklearn.decomposition import PCA

# Reduzir para 2 dimensÃµes
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)

# 2. Treinar Modelo LogÃ­stico
model_log_pca = LogisticRegression(**hyperparameters)
model_log_pca.fit(X_train_pca, y_train)

# 3. Criar Grade para VisualizaÃ§Ã£o
x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# 4. Calcular PrediÃ§Ãµes
Z = model_log_pca.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 5. Visualizar Fronteira de DecisÃ£o
plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, cmap='coolwarm', alpha=0.8)
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], 
           c=y_train, cmap='coolwarm', 
           edgecolors='k')

# Configurar GrÃ¡fico em PortuguÃªs
plt.xlabel('Primeira Componente Principal')
plt.ylabel('Segunda Componente Principal')
plt.title('Fronteira de DecisÃ£o da RegressÃ£o LogÃ­stica com PCA')
plt.colorbar(label='PrevisÃ£o')
plt.show()
```

```python
# Importar MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# Criar normalizador com intervalo [0,1]
scaler = MinMaxScaler(feature_range=(0, 1))

# Normalizar dados de treino
X_train_scaled = scaler.fit_transform(X_train)

# Normalizar dados de teste
X_test_scaled = scaler.transform(X_test)
```

```python
# 1. InstalaÃ§Ã£o e ImportaÃ§Ãµes
# !pip install tensorflow
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# 2. DefiniÃ§Ã£o do Modelo Sequencial
model = keras.Sequential([
    # Camada de Entrada
    keras.Input(shape=(12,)),  # 12 features de entrada
    
    # Primeira Camada Densa
    layers.Dense(100, activation='relu'),  # 100 neurÃ´nios, ativaÃ§Ã£o ReLU
    layers.Dropout(0.5),                   # Dropout 50% para evitar overfitting
    
    # Segunda Camada Densa
    layers.Dense(100, activation='relu'),   
    
    # Terceira Camada Densa
    layers.Dense(100, activation='relu'),
    
    # Camada de SaÃ­da
    layers.Dense(1, activation='sigmoid')   # SaÃ­da binÃ¡ria (0 ou 1)
])
```

```python
# 1. CompilaÃ§Ã£o do Modelo
model.compile(
    optimizer='adam',               # Otimizador adaptativo
    loss='binary_crossentropy',    # FunÃ§Ã£o de perda para classificaÃ§Ã£o binÃ¡ria
    metrics=['accuracy']           # MÃ©trica de avaliaÃ§Ã£o
)

# 2. Treinamento do Modelo
historico = model.fit(
    X_train,                # Dados de treino
    y_train,               # Labels de treino
    epochs=100,            # NÃºmero de iteraÃ§Ãµes
    batch_size=200,        # Tamanho do lote
    validation_split=0.2   # 20% para validaÃ§Ã£o
)

# 3. AvaliaÃ§Ã£o no Conjunto de Teste
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Perda no Teste: {test_loss:.4f}")
print(f"AcurÃ¡cia no Teste: {test_accuracy:.4f}")
```

```python
# Treinar modelo com dados normalizados
historico = model.fit(
    X_train_scaled,        # Dados de treino normalizados
    y_train,              # Labels
    epochs=100,           # NÃºmero de ciclos completos
    batch_size=200,       # Amostras por lote
    validation_split=0.2  # 20% para validaÃ§Ã£o
)
```

```python
# 1. Gerar prediÃ§Ãµes probabilÃ­sticas
y_pred_prob = model.predict(X_test_scaled)  # Retorna probabilidades [0-1]

# 2. Converter probabilidades em classes binÃ¡rias
y_pred = (y_pred_prob > 0.5).astype(int)  # 0 se prob < 0.5, 1 se prob >= 0.5

# 3. Calcular e mostrar acurÃ¡cia
accuracy = accuracy_score(y_test, y_pred)
print(f"PrecisÃ£o: {accuracy:.2f}")  # Formato: 0.XX (XX%)
```

```python
# Avaliar modelo nos dados de teste normalizados
test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test)

# Mostrar resultados em portuguÃªs
print(f"Perda no Teste: {test_loss:.4f}")
print(f"PrecisÃ£o no Teste: {test_accuracy:.4f}")
```
